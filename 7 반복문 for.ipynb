### 구구단 출력
- 3단부터 19단까지 출력해주세요
- 출력은 "2 x 2 = 4"형식입니다

```python
for i in range (3,20):
    for w in range(1,20):
        result = i * w
        print("{} x {} = {}".format(i,w,result))
    print("----------")
```

### 업다운 게임
- 컴퓨터는 1부터 100사이의 임의의 자연수를 생각합니다. 
- 플레이어는 1부터 100 사이의 임의의 자연수를 input 으로 입력합니다.
- 컴퓨터가 생각한 수가 플레이어가 입력한 수보다 크다면 Up, 작다면 Down을 출력합니다.
- 플레이어가 입력한 수가 컴퓨터가 생각한 수와 일치하면 몇번만에 맞췄는지 출력하고 프로그램을 종료합니다.

```python
import random
num = random.randint(1,100)

n = 1
while True:
    a = int(input("숫자를 입력하세요 : "))
    if num > a:
        print("UP")
        n = n+1
    elif num < a:
        print("DOWN")
        n = n+1
    else:
        print("축하합니다! {}번만에 정답을 맞췄습니다.".format(n))
        break
```
